```python
import jerry
import networkx as nx


graph = nx.grid_graph(dim=(5,5))

node_list = list(graph.nodes())

edge_list = list(graph.edges())


result = jerry.rand_kruskal(edge_list, len(node_list))


G = nx.from_edgelist(result)

# # Printing the result
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure()

nx.draw(G, pos = {x:(x[0], x[1]) for x in graph.nodes()},  node_size = 50, node_shape ='s');

plt.show()
```

```python
import random

graph = nx.grid_graph(dim=(5, 5))

# Assign random weights to the edges
for u, v in graph.edges():
    graph.edges[u, v]['weight'] = random.random()

# Compute minimum spanning tree
G= nx.minimum_spanning_tree(graph)

```

```python
plt.figure()

nx.draw(G, pos={x: (x[0], x[1])
        for x in graph.nodes()},  node_size=50, node_shape='s')

plt.show()
```
